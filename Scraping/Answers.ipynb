#### __2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.__

```python
from bs4 import BeautifulSoup
from scrapy import  Selector
import pandas as pd

data = pd.DataFrame({})
def IMDB_movies_func(*args):
    names_x = []
    release = []
    rating = []
    url = input()
    source = requests.get(url=url)
    soup = BeautifulSoup(source.content)
    names = soup.select('h3.lister-item-header > a')
    years = soup.find_all("span",class_="lister-item-year")
    Rate = soup.find_all("div",class_="ipl-rating-star small")
    values = [names,years,list(Rate)]
    for x in values:
        if x == values[0]:
            for y in x:
                names_x.append(y.text.strip())
        if x == values[1]:
            for y in x:
                release.append(y.text.strip().replace('(','').replace(')','').replace('I ',''))
        if x == values[2]:
            for y in x:
                rating.append(y.text.strip())
                
            
            
                
    #print(names_x,release)
    data['Name'] = names_x
    data['Year_of_release'] = release
    data['Rating'] = rating
    return data
```

#### __4. Write a python program to scrap book name, author name, genre and book review of any 5 books from‘www.bookpage.com’__

```python
from scrapy import  Selector
```

```python

from bs4 import BeautifulSoup
from scrapy import  Selector
import pandas as pd

def book_details(*args):
    data = pd.DataFrame({}) ''' here I difine a DataFrame'''
    source_url = input() ''' Here the function will get input url from user'''
    
    book_name = []
    author = []
    genre_list =[]
    source5 = requests.get(url=source_url)
    html = requests.get(url=source_url).content
    sel = Selector(text= html)
    genre =sel.xpath('//p[@class ="genre-links hidden-phone"]').extract()
    
    soup5 = BeautifulSoup(source5.content)
    book = soup5.select('h4.italic > a')
    author_name = soup5.find_all('p',class_="sans bold")
    
    my_list = [book,author_name,genre]
    for x in my_list:
        if x == my_list[0]:
            for y in x:
                book_name.append(y.text.strip())
                if len(book_name) == 5:
                    break
        if x == my_list[1]:
            for y in x:
                author.append(y.text.strip())
                if len(author) == 5:
                    break
        if x == my_list[2]:
            for y in x:
                genre_list.append(y.replace('<p class="genre-links hidden-phone">','').replace('\n<a href="/search?book_genre=audio">Audio</a>\n / \n<a href="/search?book_genre=romance">','').replace('</a>\n / \n<a href="/search?book_genre=contemporary_romance">','').replace('</a>\n','').replace('\n<a href="/search?book_genre=romance">','').replace('/ \n<a href="/search?book_genre=historical_romance">','').replace('</p>',''))
                if len(genre_list) == 5:
                    break
    
                
            
    data['Book_Name'] = book_name
    data['Author_Name'] = author
    data['Genre'] = genre_list
    
    return data
#class="sans"
 
```

### __5. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:__

##### __i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.__

```python
from bs4 import BeautifulSoup
from scrapy import  Selector
import pandas as pd

def cricket_detaitls_func(*args):
    data = pd.DataFrame({})
    matches_list = []
    points_list = []
    rating_list = []
    name_list =[]
    url = input()
    source = requests.get(url=url)
    soup = BeautifulSoup(source.content)
    #names = soup.select('h3.lister-item-header > a')
    matches = soup.select('tr.table-body>td:nth-child(3)')
    points = soup.find_all("td",class_="table-body__cell u-center-text")
    Rate = soup.find_all("td",class_="table-body__cell u-text-right rating")
    name = soup.find_all("span",class_="u-hide-phablet")
    
    values = [matches,points,Rate,name]
    for x in values:
        if x == values[0]:
            for y in x:
                matches_list.append(y.text.strip())
                #matches_list.sort(reverse=True)
                if len(matches_list) == 10:
                    break
        if x == values[1]:
            for y in x:
                points_list.append(y.text.strip())
                #points_list.sort(reverse=True)
                if len(points_list) == 10:
                    break
        if x == values[2]:
            for y in x:
                rating_list.append(y.text.strip())
                #rating_list.sort(reverse=True)
                if len(rating_list) == 10:
                    break
                
            
        if x == values[3]:
            for y in x:
                name_list.append(y.text.strip())
                #name_list.sort(reverse=True)
                if len(name_list) == 10:
                    break    
                
    #print(names_x,release)
    data['Matches'] = matches_list
    data['Points'] = points_list
    data['Rating'] = rating_list
    data['Country_Name'] = name_list
    return data
```

### __7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.__

```python
## creating a function using all the 4 steps
from bs4 import BeautifulSoup
from scrapy import  Selector
import pandas as pd

df_ = pd.DataFrame({})
def amazon_phone_under20k_func(*args):
    source_url = input()
    image_list = []
    rating_list =[]
    price_list = []
    name_list = []
    html = requests.get(url=source_url).content
    source_ = requests.get(url=source_url)
    sel = Selector(text= html)
    soup_= BeautifulSoup(source_.content)
    #names = soup.select('h3.lister-item-header > a')
    image_url = sel.xpath("//img[@class ='s-image']/@src").extract()
    Rate= sel.xpath("//span[@class='a-size-base']").extract()
    product_name = soup_.find_all("span",class_="a-size-base-plus a-color-base a-text-normal")
    product_price = soup_.find_all("span",class_="a-price-whole")
    values = [image_url,Rate,product_price,product_name]
    for items in values:
        if items == values[0]:
            for y in items:
                image_list.append(y.replace('',''))
                #matches_list.sort(reverse=True)
                if len(image_list) == 22:
                    break
        if items == values[1]:
            for y in items:
                rating_list.append(y.replace('<span class="a-size-base">','').replace('</span>',''))
                #points_list.sort(reverse=True)
                if len(rating_list) == 22:
                    break
        if items == values[2]:
            for y in items:
                price_list.append(y.text.strip())
                #rating_list.sort(reverse=True)
                if len(price_list) == 22:
                    break
        if items == values[3]:
            for y in items:
                name_list.append(y.text.strip())
                #rating_list.sort(reverse=True)
                if len(name_list) == 22:
                    break             
                
    #print(name_list,price_list,image_list,rating_list)
    df_['Product_Image_link'] = image_list
    df_['Product_Rating'] = rating_list
    df_['Product_Price'] = price_list
    df_['Product_Name'] = name_list
    return df_
```

### __8.   Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.__

```python

from bs4 import BeautifulSoup
import pandas as pd

data = pd.DataFrame({})
def San_Francisco_funct(*args):
    url_ = input()
    list_period =[]
    desc_list = []
    
    response = requests.get(url = url_)
    my_soup= BeautifulSoup(response.text,'html.parser')
    period_ = my_soup.select("div.forecast-label>b")
    description = my_soup.select("div.forecast-text")
    
    values = [period_,description]
    for items in values:
        if items == values[0]:
            for y in items:
                list_period.append(y.text)
    for items in values:
        if items == values[1]:
            for y in items:
                desc_list.append(y.text)  
    data['Weather_Periods'] = list_period
    data['Temperature_and_description'] = desc_list
    return data
```

### __9. Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.__
