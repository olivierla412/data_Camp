```python
## to track the progress let's import some libraries
all_details = []
from tqdm import tqdm
for i in tqdm(list_of_links):
    driver.get(i)
    product_name = driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[1]/h4[2]').text
    product_price= driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[1]/h4[1]').text
    product_reviews= driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[3]/p').text
    product_desc = driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div[2]/div/div/div[2]/div[1]/p').text
    tempj = {
    'product_name':product_name ,
    'product_price':product_price,
    'product_reviews':product_reviews,
    'product_desc':product_desc,
    'product_link':i}
    all_details.append(tempj)
```

```python
data = pd.DataFrame(all_details)
```
